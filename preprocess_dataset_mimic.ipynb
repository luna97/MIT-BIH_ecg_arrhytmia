# MIMIC labels analysis

```python
import pandas as pd
import os

file = '/media/Volume/data/MIMIC_IV/records_w_diag_icd10.csv'
exams = pd.read_csv(file)
# make study_id the index
exams.set_index('study_id', inplace=True)
# drop columns that are not needed
exams.drop(columns=['ecg_no_within_stay', 'ecg_no_within_stay', 'ecg_taken_in_hosp', 'ecg_taken_in_ed_or_hosp', 'anchor_year', 'anchor_age'], inplace=True)
exams.head()
# print the full first column 
```

```python
import json
def parse_diag(row):
    lists = [json.loads(l.replace("'", '"')) for l in row[['ed_diag_ed', 'ed_diag_hosp', 'hosp_diag_hosp', 'all_diag_hosp', 'all_diag_all']].values if l != '[]' and l != None]
    # put lists in a single list
    tortn = [item[:4] for sublist in lists for item in sublist if item.startswith('I') or item.startswith('R00')]
    # return unique values
    return json.dumps(list(set(tortn)))

parse_diag(exams.iloc[0])
```

```python
# make a set of all the unique icd10 codes in the columns ed_diag_ed, ed_diag_hosp, hosp_diag_hosp, all_diag_hosp, all_diag_all for each row
exams['icd10_codes_set'] = exams.apply(lambda row: parse_diag(row), axis=1)

# drop some columns
exams.drop(columns=['ecg_time', 'ed_stay_id', 'ed_hadm_id', 'hosp_hadm_id', 'ed_diag_ed', 'ed_diag_hosp', 'ed_diag_hosp', 'all_diag_hosp', 'all_diag_all', 'dod', ], inplace=True)
exams.head()
```

```python
# make some statisthics on the icd10 codes
icd10_codes = exams['icd10_codes_set'].explode()
# icd10_codes = icd10_codes[icd10_codes != '[]']
# add column idc10_codes_desc and idc10_codes_group initialize empty

rheumatic = ['I00', 'I01', 'I05', 'I06', 'I07', 'I08', 'I09']
hypertensive_codes = ['I10', 'I11', 'I12', 'I13', 'I15']
ischaemic_codes = ['I20', 'I21', 'I22', 'I23', 'I24', 'I25']
pulmonary_heart = ['I26', 'I27', 'I28']
cerebro_vascular = ['I60', 'I61', 'I62', 'I63', 'I64', 'I65', 'I66', 'I67', 'I68', 'I69']
arteries_diseases = ['I70', 'I71', 'I72', 'I73', 'I74', 'I77', 'I78','I79']
veins_diseases = ['I80', 'I81', 'I82', 'I83', 'I85', 'I86', 'I87', 'I88', 'I89']
hypotension = ['I95']
heart_failure = ['I50']
cardiomiopathy = ['I42', 'I43']

# common classes
first_degree_avb = ['I440']
left_bundle_bb = ['I447']
right_bundle_bb = ['I451']
bradycardia = ['R001']
tachicardia = ['I47', 'R000']
atrial_fibrillation = ['I48']



def startswith_any(codes, startswith):
    for s in startswith:
        if codes.startswith(s):
            return True
    return False

def get_group(row):
    code = row['icd10_codes_set']
    codes = json.loads(code)
    tortn = []

    for c in codes:
        if startswith_any(c, rheumatic):
            tortn.append('Rheumatic disease')
        elif startswith_any(c, hypertensive_codes):
            tortn.append('Hypertension')
        elif startswith_any(c, ischaemic_codes):
            tortn.append('Ischaemic disease')
        elif startswith_any(c, pulmonary_heart):
            tortn.append('Pulmonary Heart')
        elif startswith_any(c, cerebro_vascular):
            tortn.append('Cerebrovascular diseases')
        elif startswith_any(c, arteries_diseases):
            tortn.append('Arteries diseases')
        elif startswith_any(c, veins_diseases):
            tortn.append('Veins diseases')
        elif startswith_any(c, hypotension):
            tortn.append('Hypotension')
        elif startswith_any(c, heart_failure):
            tortn.append('Heart Failure')
        elif startswith_any(c, cardiomiopathy):
            tortn.append('Cardiomiopathy')
        # needed for labels
        elif startswith_any(c, first_degree_avb):
            tortn.append('1dAVB')
        elif startswith_any(c, atrial_fibrillation):
            tortn.append('AF')
        elif startswith_any(c, tachicardia):
            tortn.append('ST')
        elif startswith_any(c, bradycardia):
            tortn.append('SB')
        elif startswith_any(c, left_bundle_bb):
            tortn.append('LBBB')
        elif startswith_any(c, right_bundle_bb):
            tortn.append('RBBB')
        else:
            tortn.append('Other')
        return json.dumps(list(set(tortn)))

import simple_icd_10
import json

def get_desc(row):
    code = row['icd10_codes_set']
    codes = json.loads(code)
    tortn = []
    for c in codes:
        try:
            tortn.append(simple_icd_10.get_description(c[:3]) + ' (' + c[:3] + ')')
        except:
            tortn.append(c)
    return json.dumps(tortn)


exams['icd10_codes_descs'] = exams.apply(lambda row: get_desc(row), axis=1)
exams['icd10_codes_groups'] = exams.apply(lambda row: get_group(row), axis=1)
exams.head()
```

```python
# adding classes of code 15
def get_arr(row):
    if row['icd10_codes_groups'] is None:
        return []
    array = json.loads(row['icd10_codes_groups'])
    if array is None:
        return []
    return array

exams['1dAVB'] = exams.apply(lambda row: '1dAVB' in get_arr(row), axis=1)
exams['RBBB'] = exams.apply(lambda row: 'RBBB' in get_arr(row), axis=1)
exams['LBBB'] = exams.apply(lambda row: 'LBBB' in get_arr(row), axis=1)
exams['SB'] = exams.apply(lambda row: 'SB' in get_arr(row), axis=1)
exams['ST'] = exams.apply(lambda row: 'ST' in get_arr(row), axis=1)
exams['AF'] = exams.apply(lambda row: 'AF' in get_arr(row), axis=1)
```

```python
# add all the other labels
exams['Other'] = exams.apply(lambda row: 'Other' in get_arr(row), axis=1)
exams['Hypertension'] = exams.apply(lambda row: 'Hypertension' in get_arr(row), axis=1)
exams['Ischaemic disease'] = exams.apply(lambda row: 'Ischaemic disease' in get_arr(row), axis=1)
exams['Pulmonary Heart'] = exams.apply(lambda row: 'Pulmonary Heart' in get_arr(row), axis=1)
exams['Cerebrovascular diseases'] = exams.apply(lambda row: 'Cerebrovascular diseases' in get_arr(row), axis=1)
exams['Arteries diseases'] = exams.apply(lambda row: 'Arteries diseases' in get_arr(row), axis=1)
exams['Veins diseases'] = exams.apply(lambda row: 'Veins diseases' in get_arr(row), axis=1)
exams['Hypotension'] = exams.apply(lambda row: 'Hypotension' in get_arr(row), axis=1)
exams['Heart Failure'] = exams.apply(lambda row: 'Heart Failure' in get_arr(row), axis=1)
exams['Cardiomiopathy'] = exams.apply(lambda row: 'Cardiomiopathy' in get_arr(row), axis=1)
exams['Rheumatic disease'] = exams.apply(lambda row: 'Rheumatic disease' in get_arr(row), axis=1)
```

```python
import wfdb
import os
import neurokit2 as nk
import wfdb.processing as wp
import numpy as np
from pandarallel import pandarallel
pandarallel.initialize(progress_bar=True)

def analyze_ecg(row):
    """
    Analyze the ecg signal

    :param row: the row of the dataframe
    :return: the peaks
    """
    record_path = os.path.join('/media/Volume/data/MIMIC_IV/nkclean_360_12l', str(row['study_id']))
    # check path
    if not os.path.exists(record_path + '.dat'):
        print(f'Path {record_path} does not exist')
        return None
    
    record = wfdb.rdrecord(record_path)
    peaks = wp.xqrs_detect(record.p_signal[:, 1], record.fs, verbose=False)
    return peaks

exams['r_peaks'] = exams.parallel_apply(lambda row: analyze_ecg(row), axis=1)
```

```python
exams['r_peak_interval_mean'] = exams.parallel_apply(lambda row: np.mean((row['r_peaks'][1:] - row['r_peaks'][:-1]) / 360) if row['r_peaks'] is not None else None, axis=1)
exams['r_peak_variance'] = exams.parallel_apply(lambda row: np.std((row['r_peaks'][1:] - row['r_peaks'][:-1]) / 360) if row['r_peaks'] is not None else None, axis=1)
exams.head()
```
